```python
import cv2
import os
import numpy as np
```

```python
def convert_to_gray(image):
    probability = np.random.randint(0, 101)
    
    if probability <= 30:
        # Convert the image to grayscale
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        return gray_image
    else: 
        return image
def flip_image(image):
    probability = np.random.randint(0, 101)
    if probability <= 50:
        flipped_image = cv2.flip(image, 1)
        return flipped_image
    else:
        return image
        
def rotate_image(image):
    angle = np.random.randint(-7, 8)
    
    height, width = image.shape[:2]
    # Calculate the rotation matrix
    center = (width / 2, height / 2)
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1)
    # Apply the rotation to the image
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
    
    return rotated_image

#for translation
def translate_image(image):
    translated_images = []
    translate_factors = [(40, -30), (-40,30), (-40, -40)]

    for translate_factor in translate_factors:
        (h, w) = image.shape[:2]
        M = np.float32([[1, 0, translate_factor[0]], [0, 1, translate_factor[1]]])
        translated = cv2.warpAffine(image, M, (w, h))
        translated_images.append(translated)

    return translated_images
```

```python
# Path to your original dataset
dataset_dir_normal = "chest_xray/train/NORMAL"
# path to save augmented data
augmented_dataset_dir_normal = "Augmented_Data1/train/NORMAL"

# Create the save path directory if it doesn't exist
if not os.path.exists(augmented_dataset_dir_normal):
    os.makedirs(augmented_dataset_dir_normal)


for image_file in os.listdir(dataset_dir_normal):
    # Read the image
    image_path = os.path.join(dataset_dir_normal, image_file)
    image = cv2.imread(image_path)
    
    # Apply data augmentation operations
    augmented_images = []
    
    # Convert to gray
    gray = convert_to_gray(image)
    augmented_images.append(gray)
    
    # Flip image
    flipped_image = flip_image(gray)
    augmented_images.append(flipped_image)
    
    # Rotate image
    rotated_images = rotate_image(gray)
    augmented_images.append(rotated_images)
    
    # Translate image
    translated_images = translate_image(gray)
    augmented_images.extend(translated_images)
    
    # Saving the augmented images
    counter=0
    for augmented_image in augmented_images:
        # Generate a new filename for the augmented image
        augmented_filename = f"augmented_{counter}_{image_file}"
        counter += 1
        
        # Save the augmented image to the save path
        augmented_image_path = os.path.join(augmented_dataset_dir_normal, augmented_filename)
        cv2.imwrite(augmented_image_path, augmented_image)

print("Data augmentation complete for train normal data!")
# print(i)
```
