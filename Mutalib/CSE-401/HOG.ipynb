```python
import cv2
import numpy as np
from skimage.feature import hog
import matplotlib.pyplot as plt

# Load the image
img_path = 'D:/NEUB-Info/final-year-ML-projcet/chest_xray/chest_xray/train/NORMAL/IM-0115-0001.jpeg'
image = cv2.imread(img_path)

# Resize the image
resized_image = cv2.resize(image, (128, 64))

# Convert the resized image to grayscale
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

# Calculate the gradients (magnitude and angle)
gradient_x = cv2.Sobel(gray_image, cv2.CV_32F, 1, 0, ksize=3)
gradient_y = cv2.Sobel(gray_image, cv2.CV_32F, 0, 1, ksize=3)
magnitude, angle = cv2.cartToPolar(gradient_x, gradient_y)

# Visualize the magnitude and angle images
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.imshow(magnitude, cmap='gray')
ax1.set_title('Magnitude Image')
ax1.axis('off')
ax2.imshow(angle, cmap='hsv')
ax2.set_title('Angle Image')
ax2.axis('off')
plt.show()

# Divide the gradient images into cells and compute histograms
orientations = 9
pixels_per_cell = (8, 8)
cells_per_block = (3, 3)
histograms = hog(gray_image, orientations=orientations, pixels_per_cell=pixels_per_cell,
                 cells_per_block=cells_per_block, visualize=False, block_norm='L2-Hys')

# Print the HOG feature vector
print("HOG Feature Vector:")
print(histograms)
```

```python
import matplotlib.pyplot as plt
from skimage import io, feature

# Load the image
image_path = 'D:/neub-info/final-year-ML-projcet/chest_xray/chest_xray/train/NORMAL/IM-0115-0001.jpeg'
image = io.imread(image_path)

# Compute the HOG features and gradients
orientations = 9
pixels_per_cell = (8, 8)
cells_per_block = (3, 3)
hog_features, hog_image = feature.hog(image, orientations=orientations, pixels_per_cell=pixels_per_cell,
                                      cells_per_block=cells_per_block, visualize=True, multichannel=True,
                                      channel_axis=-1)

# Display the original image
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Display the HOG features
plt.subplot(1, 3, 2)
plt.imshow(hog_image, cmap='gray')
plt.title('HOG Features')

# Display the gradients
plt.subplot(1, 3, 3)
plt.quiver(hog_features[:, 1], hog_features[:, 0], color='r', width=0.0025)
plt.gca().invert_yaxis()
plt.title('Gradient Directions')

# Adjust the spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```
