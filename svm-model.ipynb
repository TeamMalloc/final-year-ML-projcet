```python
import numpy as np

# Save the features and labels to disk
# np.save('features.npy', features)
# np.save('labels.npy', labels)

# Load the features and labels from disk
features1 = np.load('/kaggle/input/svm-features/features1.npy')
labels1 = np.load('/kaggle/input/svm-features/labels1.npy')
```

```python
features2 = np.load('/kaggle/input/svm-features/features2.npy')
labels2 = np.load('/kaggle/input/svm-features/labels2.npy')

test_features1 = np.load('/kaggle/input/svm-features/test_features1.npy')
test_labels1 = np.load('/kaggle/input/svm-features/test_labels1.npy')

test_features2 = np.load('/kaggle/input/svm-features/test_features2.npy')
test_labels2 = np.load('/kaggle/input/svm-features/test_labels2.npy')
```

```python
from sklearn.svm import SVC

svm1 = SVC()

# Train SVM on the entire dataset
svm1.fit(features1, labels1)
```

```python
train_accuracy = svm1.score(features1, labels1)
print("Train Accuracy: {:.2f}%".format(train_accuracy * 100))
```

```python
test_accuracy = svm1.score(test_features1, test_labels1)
print("Test Accuracy: {:.2f}%".format(test_accuracy * 100))
```

```python
from sklearn.metrics import confusion_matrix

# # Predict labels for the training set
# train_predictions = svm1.predict(features1)

# Predict labels for the test set
test_predictions = svm1.predict(test_features1)

# Compute the confusion matrix for the training set
# train_cm = confusion_matrix(labels1, train_predictions)
# print("Confusion Matrix (Training Set):\n", train_cm)

# Compute the confusion matrix for the test set
test_cm = confusion_matrix(test_labels1, test_predictions)
print("Confusion Matrix (Test Set):\n", test_cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the confusion matrix for the test set
test_cm = confusion_matrix(test_labels1, test_predictions)

# Plot the confusion matrix as a heatmap
sns.heatmap(test_cm, annot=True, fmt="d", cmap="Blues")

# Add labels, title, and axis ticks
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.title("Confusion Matrix (Test Set)")

# Show the plot
plt.show()
```

```python
from sklearn.svm import SVC

# List of SVM kernel names
kernels = ['linear', 'poly', 'rbf', 'sigmoid']

# Train and evaluate SVM models for each kernel
for kernel in kernels:
    svm = SVC(kernel=kernel)
    svm.fit(features1, labels1)
    test_accuracy = svm.score(test_features1, test_labels1)
    print("Kernel: {}, Test Accuracy: {:.2f}%".format(kernel, test_accuracy * 100))
```
